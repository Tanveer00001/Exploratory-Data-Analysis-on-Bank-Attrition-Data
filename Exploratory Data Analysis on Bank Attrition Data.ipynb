## Analysis of numerical Data

```python
def numericalFunction(col):
    print(f"the mean of the {col} is {data[col].mean()}")
    print(f"the median of the {col} is {data[col].median()}")
    print(f"the mode of the {col} is {data[col].mode()[0]}")
    print(f"the Standard Deviation of the {col} is {data[col].std()}")
    print(f"Missing  Values in {col} are : {data[col].isnull().sum()}")
    print(f"Missing  Values in {col} are : {data[col].isnull().sum()}")
```

***standard deviation is the measure that gives how exactly our data is spread across its mean. Smaller the value better becoz iut means that most of our data is closer to mean value
